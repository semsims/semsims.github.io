```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

dfConstruction = pd.read_csv(r"C:\Users\semion.carter\Personal Documents\CTBA\CTBA\M3\ConstructionTimeSeriesdataV2.csv")

fig, ax = plt.subplots()
y1 = dfConstruction["Private Construction"]
y2 = dfConstruction["Public Construction"]
y3 = dfConstruction["Total Construction"]

x = [str(i) for i in range(len(y1))]
xlabels = ['2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014']

ax.plot(x,y1)
ax.plot(x,y2)
ax.plot(x,y3)

ax.set_xlabel("Year")
ax.set_xticklabels(xlabels)
ax.set_ylabel("Construction Spending")
ax.set_xticks(x[::12])
ax.legend(loc=1)
ax.axvspan(72,108,ymin=0,ymax=150000,facecolor='r',alpha=.1)
ax.text(90,105000,'2008 - 2010\nHousing Collapse',fontsize = '18',verticalalignment='center',horizontalalignment='center')

fig.set_size_inches(14,6)
fig.suptitle("Construction Costs Across Private and Public Sector", fontsize='18')
```
